### Creating a Conda Environment From Scratch 

Make sure you have [conda installed](https://conda.io/docs/user-guide/install/). Conda's documentation also contains a step by step guide on how to [create and manage conda environments](https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html)

Here we are going to outline three basic and simple steps to get started: 

1. Creating and environment 
2. Activating the environment and finally 
3. Installing Geospatial packages into the environment 

#### Create an environment 

On your machine (assuming we're all using Windows), hit the _windows key_ and search for `Anaconda Prompt`. Run the program as administrator to get full acess and control. 

**Take Note**: The prompt will open with the default `base` environment already activated 

In the console type the following command: 

`conda create --name hsvg245`

In this case `hsvg245` is the name of our environment. You can name it whatever you want. 

Below is a print out of the console 



Now activating the environment type...

`conda activate hsvg245` 

You should now see the name of the environment at the command line prompt (at the beginning of the line). 



Now we can **install packages** into the activated environment: 

```
# Install jupyter lab + git extension
conda install -c conda-forge jupyterlab jupyterlab-git

# Install packages
conda install -c conda-forge geopandas
conda install -c conda-forge matplotlib
conda install -c conda-forge geojson
conda install -c conda-forge folium
# ... install other packages
```

After you have installed all required packages, you can start working in a local JupyterLab environment that is linked to your `hsvg245` conda environment:

`jupyter lab`
